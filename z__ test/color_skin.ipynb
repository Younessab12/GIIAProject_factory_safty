```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Load an image
image = cv2.imread('img.jpg')
plt.imshow(image)
```

```python
# Convert the image from BGR to YCrCb
ycrcb_image = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
plt.imshow(ycrcb_image)
```

```python
# Define the range for skin color
# These values can vary; you might need to adjust them
min_YCrCb = np.array([0, 133, 77], np.uint8)
max_YCrCb = np.array([255, 173, 127], np.uint8)

# Find skin region
skinRegion = cv2.inRange(ycrcb_image, min_YCrCb, max_YCrCb)

# Do a bit-wise and with the original image
skin = cv2.bitwise_and(image, image, mask = skinRegion)
plt.imshow(skin)
```

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('path_to_image.jpg')

# Convert to YCrCb color space
ycrcb_image = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)

# Define the range for skin color
# These values can vary; you might need to adjust them
min_YCrCb = np.array([0, 133, 77], np.uint8)
max_YCrCb = np.array([255, 173, 127], np.uint8)

# Find skin region
skinRegion = cv2.inRange(ycrcb_image, min_YCrCb, max_YCrCb)

# Do a bit-wise and with the original image
skin = cv2.bitwise_and(image, image, mask = skinRegion)

# Display the image (optional)
cv2.imshow("Skin", np.hstack([image, skin]))

# Wait for key press
cv2.waitKey(0)
cv2.destroyAllWindows()
```
